```python
def speed_cost(speed):
    speedLimit = 50.0
    optimalSpeed = 49.5
    target_v = optimalSpeed;
    k = 1.0 / target_v;
    d = k * target_v;
    
    if speed < target_v:
        return (-0.8 / target_v) * speed + 0.8;

    if speed == target_v:
        return 0.0;

    if target_v < speed and speed <= speedLimit:
        return (-1.0 / target_v) * speed + 1.0;

    return 1.0;
```

```python
speed = np.arange(0.0, 55.0, 0.5)
costs = [speed_cost(s) for s in speed]

plt.figure(figsize=(18,12))
plt.plot(speed, costs, color='green', linestyle='solid', marker='o', markerfacecolor='blue', markersize=6)
```
